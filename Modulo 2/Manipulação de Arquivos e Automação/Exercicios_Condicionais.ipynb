Exercício 4 - Calculadora de Desconto. 💵 🤑

Uma loja dá 10% de desconto para compras acima de R$ 100,00. Faça um programa que:

1º - Pergunta o valor da compra.

2º - Calcula o valor final com desconto (se aplicável).

3º - Mostra o valor a pagar.

```python
### Seu código aqui
compra = int(input('qual foi o valor da compra?'))
if compra > 100:
    desconto = compra *0.10
    print(f' sua compra foi de {compra} e teve um desconto de {desconto}, sendo assim o valor a pagar é de {compra - desconto}')
else:
    print(f'sua compra foi menor do que o necessario para o desconto, pague o valor inteiro!!! {compra}')
```

Exercício 5 - Classificação de Notas 🧑‍🎓

Pergunte a nota de um aluno e dependendo da nota que ele informar, o classifique de acordo com as regras abaixo:

Classifique a nota do aluno:

- 9 até 10: "Excelente"

- 7 até 8.9: "Bom"

- 5 até 6.9: "Regular"

- 0 até 4.9: "Reprovado"

mostre para o aluno a classificação dele através de um print, exemplo print("Excelente").

Exercício 6 - Acesso ao Sistema 🔑

Crie um login simples:

- usuario_correto = "admin"

**Peça para o usuário digitar primeiro o login e depois a senha, caso login e senha estejam corretos programa deve mostrar "Acesso permitido" se não "Acesso negado"**


```python
usuario_correto = "admin"
senha_correta = 1234
usuario = input('Qual é o usuario?')
senha = int(input('Qual é a senha?'))
if usuario == usuario_correto and senha == senha_correta:
    print('Acesso permitido')
else:
    print('Acesso negado')



```

Exercício 7 - Maior Número 📉

Peça 3 números e mostre qual é o maior deles.

```python
n1 = int(input('Qual o primeiro numero?'))
n2 = int(input('Qual o segundo numero?'))
n3 = int(input('Qual o terceiro numero?'))

if n1>n2 and n1>n3:
    print(f'{n1} é o maior numero ')
elif n2>n1 and n2>n3:
    print(f'{n2} é o maior numero ')
else:
    print(f'{n3} é o maior numero ')
```

Exercício 8 - Dia da Semana
Peça um número de 1 a 7 e mostre o dia da semana correspondente: 📅

- 1: Domingo

- 2: Segunda

- ... até 7: Sábado

```python
numero = int(input("Digite um número de 1 até 7"))
if numero == 1:
    print('domingo')
elif numero == 2:
    print('segunda')
elif numero == 3:
    print('terça')
elif numero == 4:
    print('quarta')
elif numero == 5:
    print('quinta')
elif numero == 6:
    print('sexta')
else:
    print('sabado')


```

Exercício 9 - Calculadora Simples 📏📐

- Faça uma calculadora que:

- Pede dois números

- Pede a operação (+, -, *, /)

- Dependendo da operação mostre o resultado.



```python
n1 = float(input('me diga o primeiro numero'))
n2 = float(input('me diga o segundo numero'))
operacao = input('digite a operaçao (+,-,*,/)')

if operacao =="+":
    print(f'o resultado é {n1+n2}')
elif operacao == '-':
    print(f'o resultado é {n1-n1}')
elif operacao == '*':
    print(f'o resultado é {n1*n2}')
else:
    print(f'o resultado é {n1/n2}')
```

Exercício 10 - Sistema de Multas 🚙🚘🧾


- Velocidade máxima: 80 km/h 🚫

- Até 80: "Sem multa"

- 81-100: "Multa leve"

- Acima de 100: "Multa grave"

```python
max = float(input('Qual a velocidade do carro?'))
if max<=80:
    print('sem multa')
elif max>80 and max<=100:
    print('Multa leve')
else:
    print('multa grave')
```
